```python
# Load needed Python modules
import netCDF4
import matplotlib.pyplot as plt
import os
from netCDF4 import num2date

# Switch to directory containing file(s)    
# os.chdir('/Volumes/Samsung_T5/plot-data-1')
fn='%s/%s'%('/nfs/cnhlab003/cnh/mur-sst/','20130109090000-JPL-L4_GHRSST-SSTfnd-MUR-GLOB-v02.0-fv04.1.nc')
# Create "handle" to access netCDF file
tFile = netCDF4.Dataset(fn)


# Extract data from file
sst=tFile['analysed_sst']
date_num = tFile['time'][:]
units = tFile.variables['time'].units

# Check dimensions
print(sst.shape)

#Print time
date = num2date(date_num, units)[0]
print(date)

# Extract a subregion
sstReg=sst[0,11300:14000,1750:2750]-273.


# Make a very basic plot
fig = plt.imshow(sstReg,origin='lower',cmap='prism');plt.colorbar();plt.show()
#plt.imsave(str(date) + '.png', sstReg, origin = 'lower', cmap = 'prism')
```
